```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv ('train.csv')
```

```python
df.info()
```

```python
# Plotting fitur NAME_INCOME_TYPE terhadap fitur target
# Plotting fitur NAME_INCOME_TYPE terhadap fitur target
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='NAME_INCOME_TYPE', hue='TARGET', data=df)

# Menambahkan label di atas setiap bar
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2, height + 1000, f'{height}', ha='center')

plt.title('NAME_INCOME_TYPE vs Target')
plt.xlabel('Income Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Plotting fitur Gender_Map terhadap fitur target
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender_Map', hue='TARGET', data=df)
plt.title('CODE_GENDER vs Target')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot distribusi EXT_SOURCE_2 dan EXT_SOURCE_3
plt.figure(figsize=(10, 6))

# Plot distribusi EXT_SOURCE_2
plt.subplot(1, 2, 1)
sns.histplot(data=df, x='EXT_SOURCE_2', bins=30, kde=True, color='blue')
plt.title('Distribution of EXT_SOURCE_2')

# Plot distribusi EXT_SOURCE_3
plt.subplot(1, 2, 2)
sns.histplot(data=df, x='EXT_SOURCE_3', bins=30, kde=True, color='green')
plt.title('Distribution of EXT_SOURCE_3')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Konversi DAYS_BIRTH ke tahun dan simpan dalam kolom baru 'AGE'
df['AGE'] = df['DAYS_BIRTH'] // -365

# Plot histogram usia pelanggan
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='AGE', bins=20, kde=True, color='orange')
plt.title('Distribution of Customer Age')
plt.xlabel('Age (years)')
plt.ylabel('Frequency')
plt.show()
```
